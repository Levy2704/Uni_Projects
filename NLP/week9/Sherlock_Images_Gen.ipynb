```python
# Get the set of English stopwords from NLTK
stopwords = set(nltk_stopwords.words('english'))

def masked_wordcloud(text, mask, stopwords):
    wordcloud = WordCloud(background_color='white',
                        stopwords=stopwords,
                        max_words=15000,
                        max_font_size=86,
                        random_state=42,
                        mask=mask)
    wordcloud.generate(text)
    figure_size = (24.0, 16.0)
    plt.figure(figsize=figure_size)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()  
list(stopwords)[:15]
```

```python
text=open("C:\\Users\\josep\\Documents\\Torrens_University\\NLP\\week9\\sherlock\\fina.txt").read()
mask = np.array(Image.open('C:\\Users\\josep\\Documents\\Torrens_University\\NLP\\week9\\sherlock\\sherlock-holmes-and-watson.jpg'))
masked_wordcloud(text, mask,stopwords)
```
